```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import os
import time
import numpy as np

# --- Task 1a ---
# Load the image using OpenCV
image_path = 'img.jpg'  # Replace with your image path
image = cv2.imread(image_path)

# Get dimensions and number of channels
dimensions = image.shape
height, width, channels = dimensions

# Display the image using Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display
plt.imshow(image_rgb)
plt.title(f"Dimensions: {dimensions}")
plt.show()

# Output the dimensions
print(f"Width: {width} pixels")
print(f"Height: {height} pixels")
print(f"Number of channels: {channels}")

# --- Task 1b ---
# Load the image using PIL to get metadata
img = Image.open(image_path)

# Get size, height, width
size = os.path.getsize(image_path)  # Size in bytes
width, height = img.size

# Get creation/modification date
timestamp = os.path.getmtime(image_path)  # Last modified time
date = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(timestamp))

# Get image mode (Coding Method) and bit depth
mode = img.mode
bit_depth = img.bits if hasattr(img, 'bits') else 8  # Some formats don't have the 'bits' attribute

# Output metadata
print(f"Size: {size / 1024:.2f} KB")  # Convert bytes to KB
print(f"Date: {date}")
print(f"Image Mode (Coding Method): {mode}")
print(f"Bit Depth: {bit_depth} bits")
print(f"Height: {height} pixels")
print(f"Width: {width} pixels")

# --- Task 1c ---
# Example double values outside [0, 255]
values = np.array([300.0, -20.0, 260.0, 0.0, 255.0, 128.0])

# Convert to uint8
uint8_values = values.astype(np.uint8)

# Output conversion results
print("Original Values:", values)
print("After Conversion to uint8:", uint8_values)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = 'img.jpg'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Binarize the image using threshold
threshold_value = 127
_, binarized_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Display both original and binarized images using Matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Binarized image
plt.subplot(1, 2, 2)
plt.imshow(binarized_image, cmap='gray')
plt.title('Binarized Image (Threshold = 127)')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = 'img.jpg'  # Replace with your image path
image = cv2.imread(image_path)

# Check the number of channels and size of the image
if len(image.shape) == 3:
    # Three-channel image (Color)
    height, width, channels = image.shape
    print(f"The image is a three-channel image with dimensions: {width}x{height} and {channels} channels.")

    # Convert to grayscale using cv2.cvtColor
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Binarize the grayscale image
    threshold_value = 127
    _, binarized_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

    # Display the original, grayscale, and binarized images
    plt.figure(figsize=(15, 5))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
    plt.title('Original Image')
    plt.axis('off')

    # Grayscale image
    plt.subplot(1, 3, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')

    # Binarized image
    plt.subplot(1, 3, 3)
    plt.imshow(binarized_image, cmap='gray')
    plt.title('Binarized Image (Threshold = 127)')
    plt.axis('off')

    plt.show()

elif len(image.shape) == 2:
    # One-channel image (Grayscale)
    height, width = image.shape
    print(f"The image is a one-channel image with dimensions: {width}x{height}.")

    # Display the grayscale image
    plt.imshow(image, cmap='gray')
    plt.title('One-Channel Grayscale Image')
    plt.axis('off')
    plt.show()
else:
    print("The image format is unsupported.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = 'img.jpg'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Set the threshold value
threshold_value = 127

# Get the image dimensions
height, width = image.shape

# Create an empty output image for the binarized result
binarized_image = np.zeros((height, width), dtype=np.uint8)

# Manual binarization using nested loops
for i in range(height):
    for j in range(width):
        if image[i, j] > threshold_value:
            binarized_image[i, j] = 255  # Set to white if above threshold
        else:
            binarized_image[i, j] = 0    # Set to black if below threshold

# Display both original and binarized images using Matplotlib
plt.figure(figsize=(10, 5))

# Original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

# Binarized image
plt.subplot(1, 2, 2)
plt.imshow(binarized_image, cmap='gray')
plt.title(f'Binarized Image (Threshold = {threshold_value})')
plt.axis('off')

plt.show()
```
